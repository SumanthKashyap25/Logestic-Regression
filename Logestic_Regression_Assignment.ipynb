```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
bank=pd.read_csv("bank-full.csv",sep=';')
bank
```

**EDA**

```python
bank.info()
```

```python
#categorical variables had to be converted to numerical with the help of One Hot Encoding
bank1=pd.get_dummies(bank,columns=['job','marital','education','contact','poutcome'])
bank1
```

```python
#many columns are hidden so to view all the columns
pd.set_option("display.max.columns",None)
bank1
```

```python
bank1.info()
```

```python
#Binary Encoding
bank1["default"]=np.where(bank1["default"].str.contains("yes"),1,0)
bank1["housing"]=np.where(bank1["housing"].str.contains("yes"),1,0)
bank1["loan"]=np.where(bank1["loan"].str.contains("yes"),1,0)
bank1["y"]=np.where(bank1["y"].str.contains("yes"),1,0)
bank1
```

```python
bank1["month"].value_counts()
```

```python
month_order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
```

```python
bank1=bank1.replace(month_order)
bank1
```

```python
bank1.info()
```

**Model Creation**

```python
#defining dependant and independant variables
x=pd.concat([bank1.iloc[:,0:11],bank1.iloc[:,12:]],axis=1)
y=bank1.iloc[:,11]
```

```python
#Logistic Regression & fit the model
classifier=LogisticRegression()
classifier.fit(x,y)
```

**Prediction of Model**

```python
#Predict for x dataset
y_predict=classifier.predict(x)
y_predict
```

```python
y_predict_df=pd.DataFrame({'Actual Y':y,'Predicted Y':y_predict})
y_predict_df
```

**Model Accuracy**

```python
#Confusion Matrix for model accuracy
confusion_matrix=confusion_matrix(y,y_predict)
confusion_matrix
```

```python
#As rhe accuracy is 0.89221 which is greater than 0.5 so, [:,1] threshold values>0.5=1 else [:,0] theshold values<0.5=0
classifier.predict_proba(x)[:,1]
```

```python
#Plotting the ROC curve and finding AUC score
fpr,tpr,threshold=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='Red')
auc_score=roc_auc_score(y,y_predict)

plt.plot(fpr,tpr,color='Red',label='logit model(area=%0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.title("ROC Curve")
plt.xlabel('False Positive Rate or (1-True Negative Rate)')
plt.ylabel("True Positive Rate")
plt.legend()
plt.savefig('ROC',dpi=300)
plt.show()
auc_score

print('AUC Accuracy',auc)
```
